```python
import cppapi
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import log_loss
from scipy.special import softmax


enc = OneHotEncoder(sparse_output=False)

X = np.random.randint(0, 3, (100, 1))

Xenc = enc.fit_transform(X)

print(Xenc)
```

```python
loss = cppapi.LossFn(cppapi.LossType.CROSS_ENTROPY)


Q = np.random.randn(100, 3)
print("Predictions:\n", Q)
print("Python softmax:\n", softmax(Q, axis=1))

cppapi.loss_functions.forward(loss, Q.astype(np.float32), Xenc.astype(np.float32))

print("softmax matrix\n", loss.place_holder)

print("cpp loss value:", loss.loss_value)

print(log_loss(Xenc, softmax(Q, axis=1)))

np.sum(-Xenc * np.log(loss.place_holder) - (1-Xenc) * np.log(1-loss.place_holder)) / 100
```
